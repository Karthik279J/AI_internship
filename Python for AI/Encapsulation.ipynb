```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.__age = age

    def get_age(self):
        return self.__age
    def set_age(self,age):
        if age > 0:
            self.__age = age
        else:
            print("Age must be positive")
p1 = person("Agnes",21)
print(p1.name)
print(p1.get_age())
p1.set_age(26)
print(p1.get_age())
```

```python
class student:
    def __init__(self,name):
        self.name = name
        self.grade = 0
    def set_grade(self,grade):
        if 0<=grade <= 100:
            self.__grade = grade
        else:
            print("Grade must be between 0 and 100")
    def get_grade(self):
        return self.__grade
    def get_status(self):
        if self.__grade >= 60:
            return "passed"
        else:
            return "Failed"
Student = student("Agnes")
Student.set_grade(85)
print(Student.get_grade())
print(Student.get_status())
    
```

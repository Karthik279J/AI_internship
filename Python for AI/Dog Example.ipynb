```python
class Dog:
    """A sample dog class to demonstrate classes and objects"""

    # constructor/initializer
    def __init__(self, name, age, breed):
        self.name = name          # dog's name
        self.age = age            # dog's age
        self.breed = breed        # dog's breed
        self.is_hungry = True     # default state (boolean)

    # method: makes the dog bark
    def bark(self):
        return f"{self.name} says: boww boww!"

    # method: feed the dog (changes the internal state)
    def feed(self):
        if self.is_hungry:
            self.is_hungry = False
            return f"You fed {self.name}. {self.name} is no longer hungry."
        else:
            return f"{self.name} is not hungry right now."

    # method: makes the dog celebrate its birthday (changes age)
    def celebrate_birthday(self):
        self.age += 1
        return f"Happy birthday, {self.name}! Now {self.name} is {self.age} years old."

    # special method: string representation of the object
    def __str__(self):
        hunger = "hungry" if self.is_hungry else "not hungry"
        return f"{self.name} ({self.breed}), Age: {self.age}, {hunger}"


# creating objects from class
if __name__ == "__main__":
    dog1 = Dog(name="Jack", age=1, breed="Pomerian")
    dog2 = Dog(name="Puppy", age=2, breed="Pomerian")

    print(dog1)
    print(dog2)
    print(dog1.bark())
    print(dog2.feed())
    print(dog2.feed())
    print(dog1.celebrate_birthday())
    print(dog1)
```

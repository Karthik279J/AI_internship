```python
sns.countplot(x='Heart Disease', data=df)
plt.title("Heart Disease Distribution")
plt.show()
```

```python
sns.boxplot(x='Heart Disease' , y='Cholesterol', data=df)
plt.title=("Cholesterol nvs Heart Disease")
plt.show()
```

```python
Insights:
1.Patients with heart disease tend to have higher age, cholesterol, and ST depression.
2.Exercise-induced angina is more common among heart disease patients.
3.Abnormal thallium test results and higher vessel count strongly indicate heart disease.
```

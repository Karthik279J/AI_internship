```python
def add(a, b):
    result = a + b
    return result

x = add(6, 5)
print(x)
```

```python
def rectangle_area(length, width):
    area = length * width
    return area

print(rectangle_area(5, 6))
```

```python
def greet(name="friend"):
    print(f"hello, {name}")

greet("alice")
greet()
```

```python
def introduce(name, age, city):
    print(f"I'm {name}, my age is {age}, I'm from {city}")

introduce(name="Agnes", age=25, city="Vijayawada")
```
